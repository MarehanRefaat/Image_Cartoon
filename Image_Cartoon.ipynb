```
import cv2 #for image processing
import numpy as np #to store image
import imageio #to read image stored at particular path

import sys
import matplotlib.pyplot as plt
import os
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
```

```
img = cv2.imread('/content/drive/MyDrive/Sample_photo.jpg')
image2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
plt.imshow(image2)
```

```
gray_image = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
```

```
plt.imshow(gray_image,cmap='gray')
```

```
smooth_gray = cv2.medianBlur(gray_image,5) #have to be odd number 
plt.imshow(smooth_gray,cmap='gray')
```

```
get_Edge = cv2.adaptiveThreshold(smooth_gray, 255, 
  cv2.ADAPTIVE_THRESH_MEAN_C, 
  cv2.THRESH_BINARY, 11,2)
plt.imshow(get_Edge,cmap='gray')
```

```
color_image = cv2.bilateralFilter(image2, 9, 300, 300)
plt.imshow(color_image,cmap='gray')
```

```
cartoon_image = cv2.bitwise_and(color_image, color_image, mask=get_Edge)
plt.imshow(cartoon_image,cmap='gray')
```

```

f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(np.rot180(image2,2))
f.add_subplot(1,2, 2)
plt.imshow(np.rot180(cartoon_image,2))
plt.show(block=True)
```

```
f, axarr = plt.subplots(1,2)
axarr[0].imshow(image2)
axarr[0].set_title('Orginal Image')
axarr[1].imshow(cartoon_image)
axarr[1].set_title('Cartoon Image')
```
